## Read the File into a pandas DataFrame

```python
import pandas as pd
## Load the CSV into a DataFrame
df = pd.read_csv('female_olympic_swimmers.csv')
# Preview the first few rows
df.head()
```

## Filter Data by Year & Count by Country Code(NOC)

```python
# Filter for years 2000 to 2016
df_filtered = df[(df['Year'] >= 2000) & (df['Year'] <= 2016)]

# Count swimmers by country (NOC)
country_counts = df_filtered['NOC'].value_counts()

# Display the top country
top_country = country_counts.idxmax()
print("Country with most female Olympic swimmers:", top_country)
```

## Identifying the Tallest Medalist-2016 Olympic Games

```python
# Filter for 2016 competitors
swimmers_2016 = df[df['Year'] == 2016]

# Keep only medalists
medalists_2016 = swimmers_2016[swimmers_2016['Medal'].notnull()]

# Find the tallest medalist
tallest_medalist = medalists_2016.loc[medalists_2016['Height'].idxmax()]

# Display her full name
print("Tallest medalist in 2016:", tallest_medalist['Name'])
```

## Strategy Team:Highest Paid Athletes Analysis

## Load the Dataset from GitHub

```python
import pandas as pd

# Load the CSV directly from GitHub
url = 'https://raw.githubusercontent.com/ali-ce/datasets/master/Most-paid-athletes/Athletes.csv'
df_athletes = pd.read_csv(url)

# Preview the data
df_athletes.head()
```

## Cleaning and Converting Total Pay Column

```python
# Strip "$" and "," characters from the 'Total Pay' column
df_athletes['Total Pay'] = df_athletes['Total Pay'].str.replace('$', '', regex=False)
df_athletes['Total Pay'] = df_athletes['Total Pay'].str.replace(',', '', regex=False)

# Convert to numeric data type
df_athletes['Total Pay'] = pd.to_numeric(df_athletes['Total Pay'])

# Preview the cleaned column
df_athletes['Total Pay'].head()
```

## Correlation Between Height and Total Pay

```python
# Check Columns name
print(df_athletes.columns)
```

```python
# Convert 'Height (cm)' to numeric
df_athletes['Height (cm)'] = pd.to_numeric(df_athletes['Height (cm)'], errors='coerce')

# Drop rows with missing values in relevant columns
correlation_data = df_athletes[['Height (cm)', 'Total Pay']].dropna()

# Calculate the correlation
correlation = correlation_data.corr().loc['Height (cm)', 'Total Pay']

# Round to the nearest hundredths place
rounded_correlation = round(correlation, 2)
print("Correlation between Height and Total Pay:", rounded_correlation)
```

## Age Range of Top-Paid Atheletes

```python
# Convert 'Year of birth' to numeric (if not already)
df_athletes['Year of birth'] = pd.to_numeric(df_athletes['Year of birth'], errors='coerce')

# Assume the dataset year is 2023 (as per instructions) to calculate current age
df_athletes['Age'] = 2023 - df_athletes['Year of birth']

# Drop any rows with missing ages
age_data = df_athletes['Age'].dropna()

# Calculate percentiles
q75 = age_data.quantile(0.75)
q25 = age_data.quantile(0.25)

# Calculate IQR and round
iqr = round(q75 - q25, 2)
print("Interquartile Range (Age):", iqr)
```

## Mean Total Pay by Sport

```python
# Group by sport and calculate mean total pay
mean_pay_by_sport = df_athletes.groupby('Sport')['Total Pay'].mean().sort_values(ascending=False)

# Display the sport with the highest mean pay
top_sport = mean_pay_by_sport.idxmax()
top_pay = mean_pay_by_sport.max()

print("Sport with Highest Mean Pay:", top_sport)
print("Average Pay:", round(top_pay, 2))

# Optionally display the full list
print("\n Full Mean Pay by Sport:")
print(mean_pay_by_sport.round(2))
```

## Identify Top Defensive Underused Players

## Create New Features

```python
# Calculate 'stocks_per_game' (steals + blocks) divided by games played
df['stocks_per_game'] = (df['steals'] + df['blocks']) / df['games']

# Calculate 'minutes_per_game'
df['minutes_per_game'] = df['minutesPg'] / df['games']
```

```python
# Convert columns to numeric (invalid entries will be set to NaN)
df['steals'] = pd.to_numeric(df['steals'], errors='coerce')
df['blocks'] = pd.to_numeric(df['blocks'], errors='coerce')
df['games'] = pd.to_numeric(df['games'], errors='coerce')
df['minutesPg'] = pd.to_numeric(df['minutesPg'], errors='coerce')
```

```python
# Confirm all columns are numeric and have no missing data
print(df[['steals', 'blocks', 'games', 'minutesPg']].info())
```

## Filter Underused Players & Rank by Defensive Value

```python
# Filter underused players: less than 25 minutes per game
underused_df = df[df['minutes_per_game'] < 25]

# Get top 10 underused players by stocks per game
top_defenders = underused_df.sort_values(by='stocks_per_game', ascending=False).head(10)

# Display their names and metrics
top_defenders[['playerName', 'team', 'minutes_per_game', 'stocks_per_game']]
```

```python
# Underused player with the highest "Stocks per Game"
top_player = top_defenders.iloc[0]['playerName']
print(f"Underused player with highest Stocks per Game: {top_player}")
```
